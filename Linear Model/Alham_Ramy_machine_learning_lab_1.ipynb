```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pandas import read_csv
def load_data(filepath: str) -> pd.DataFrame:
#define a function that takes the file path, and then return the data.
  column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']#gives names to column
  data = read_csv(filepath, header=None, skiprows=22, names=column_names, sep="\s+")#read the file and skip the first 22 rows, and give the name of the cloumn
  data1=pd.DataFrame(data=data)
  for i in range(len(data1)-1):
      if np.isnan(data1["B"][i]): # Test element-wise for NaN if the B==NaN replce B with CRIM in the next row, and LSTAT with ZN of the next row, and MEDV with INDUS of the next row.
        data1.B[i]=data1.CRIM[i+1]
        data1.LSTAT[i]=data1.ZN[i+1]
        data1.MEDV[i]=data1.INDUS[i+1]
  data1 = data1.dropna() #Drop the rows where at least one element is missing.
  data1.reset_index(drop=True, inplace=True)# Reset the raw index

  return data1

data1 = load_data('boston.csv')
data1.head()#show the head of the data
```

```python
import random
x = data1.RM# i will use the RM data as x, where  RM is average number of rooms per dwelling
y = data1.MEDV# i will use the MEDV data as y, where  MEDV is Median value of owner-occupied homes in $1000's

def lm(x, y) :
#define a function that takes x and y as a list and returns the fit line parameter m and b
#m -- slope of the line
#b -- y intercept
    m=random.randint(0,10)
    b=random.randint(0,10)
    return m, b   
m,b=lm(x,y)#call the function.
y_pred= m*x+b# calcualte the y prediction
plt.scatter(x, y) #Visualise the linear model returned from lm of the scatter plot of the input and target data.
plt.title("RM and MEDV ")
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.plot(x, y_pred)
plt.show()
```

```python
#mean absolute error---Evaluating our linear model.
def mae(y, y_pred) :
  sum=0
  for i in range(len(y)):
    sum += abs(y[i]- y_pred[i]) 
  error=sum/len(y)
  return error
error=mae(y,y_pred)
print(error)
```

```python
#another method to get the mean absolute error

def mae(y, y_pred):
    y, y_pred = np.array(y), np.array(y_pred)
    return np.mean(np.abs(y - y_pred))
    
print(mae(y, y_pred))
```

```python
#least-squares regression method--to get the optimal m and b.
def lm(x, y):
  xy = []
  for i, val in enumerate(x):
      xy.append(x[i] * y[i]) #make a list of xy.
      
  x_sqrt = [i**2 for i in x]#make a list of s sqaure
  
  n  = len(x)
  m = (n*np.sum(xy) - np.sum(x)*np.sum(y)) / (n*np.sum(x_sqrt) - np.sum(x)**2)
  b = (np.sum(y) - m*np.sum(x))/n
  return m, b
```

```python
m,b=lm(x,y)#call the function
u=x*m+b#we will calculate y pred which y prediction = u = mx+b.
plt.scatter(x,y)
plt.title("RM and MEDV ")
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.plot(x,u, color='r')
plt.show()
```

```python
#call the functin mae to calculate mean absolute error for prediction after applying the least-squares regression method, we see the mean square error now is a little.    
print(mae(y, u))
```
